```matlab
figure(1); clf reset;

% Ymed is median inhabited latitude
plot(obis.Ymed, obis.CTmax, 'r*');
hold on; 

% choose Thab at Imax of inhabited Temp percentiles
Imax = find(obis.prct == 95); % Index for 95th percentile
plot(obis.Ymed, obis.Tprct(:, Imax), 'b*'); 

plot(obis.Ymed, obis.CTmax - obis.Tprct(:, Imax), 'g*');

legend('CTmax', 'Thab (95%)', 'TSM')

plot([-80, 80], [0, 0], 'k:')
xlabel('Latitude');
ylabel('Temperature [C]');
```
